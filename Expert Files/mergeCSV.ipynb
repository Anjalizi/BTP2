```python
data = pd.read_csv('Q1-MFN6 - Sheet1.csv')
data.head()
```

```python
with open('heights-bmi.csv', 'w') as file:
    writer = csv.writer(file)
    for i in range(50):
        row_to_add = []
        height = data.loc[i]['Height']/100
        weight = data.loc[i]['Weight']
        bmi = weight / (height * height)
        row_to_add.append(height*height)
        row_to_add.append(bmi)
        writer.writerow(row_to_add)
```
