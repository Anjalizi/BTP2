```python
import mne
import numpy as np
import csv
```

```python
from scipy.stats import skew
```

```python
def pooled_variance(df):
    variance_vals = np.var(df)
    pooled_var = 0
    sample_size = len(df)
    num_channels = df.shape[1]

    for val in variance_vals:
        pooled_var += (sample_size - 1) * val
    pooled_var = pooled_var/((sample_size*num_channels) - num_channels)
    return pooled_var
```

```python
with open('Variance Values/eeg9-skew.csv', 'w') as file:
    writer = csv.writer(file)
    for row in range(20):
        row_to_add = []
        data = mne.io.read_raw_edf('Splitted Files/eeg_9/eeg9_reduced{}.edf'.format(row+1), preload=True)
        df = data.to_data_frame()
        pooled_var = pooled_variance(df)
        row_to_add.append(pooled_var)  
        writer.writerow(row_to_add)
```
