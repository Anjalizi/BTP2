```python
import pandas as pd
```

```python
data = pd.read_csv('Combined-780 - Sheet1.csv')
data
```

```python
X = data.drop(columns=['Signal','Label'])
y = data['Label']
```

```python
import numpy as np
np.isnan(X).any()
```

```python
X.loc[[2,3,4,50]]
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.model_selection import KFold
from sklearn import svm
from sklearn import preprocessing as pre

kf = KFold(n_splits=7)
acc = []
sens = []
spec = []
for train_index, test_index in kf.split(X,y):
    # print("%s %s" % (train_index, test_index))
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    X_train_scaled = pre.scale(X_train)
    X_test_scaled = pre.scale(X_test)
    model = svm.SVC(kernel='linear',C=20).fit(X_train_scaled, y_train)
    score = model.score(X_test_scaled, y_test)
    y_pred = model.predict(X_test_scaled)
    cm1 = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix : \n', cm1)

    total1=sum(sum(cm1))
    #####from confusion matrix calculate accuracy
    accuracy1=(cm1[0,0]+cm1[1,1])/total1
    print ('Accuracy : ', accuracy1)

    sensitivity1 = cm1[0,0]/(cm1[0,0]+cm1[0,1])
    print('Sensitivity : ', sensitivity1 )

    specificity1 = cm1[1,1]/(cm1[1,0]+cm1[1,1])
    print('Specificity : ', specificity1)
    acc.append(accuracy1)
    sens.append(sensitivity1)
    spec.append(specificity1)
    
print(max(acc))
```

```python
from sklearn.neighbors import KNeighborsClassifier

kf = KFold(n_splits=8)
acc = []
for train_index, test_index in kf.split(X,y):
    # print("%s %s" % (train_index, test_index))
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    X_train_scaled = pre.scale(X_train)
    X_test_scaled = pre.scale(X_test)
    knn = KNeighborsClassifier(n_neighbors=5).fit(X_train_scaled, y_train)
    score = knn.score(X_test_scaled, y_test)
    y_pred = knn.predict(X_test_scaled)
    cm1 = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix : \n', cm1)

    total1=sum(sum(cm1))
    #####from confusion matrix calculate accuracy
    accuracy1=(cm1[0,0]+cm1[1,1])/total1
    print ('Accuracy : ', accuracy1)

    sensitivity1 = cm1[0,0]/(cm1[0,0]+cm1[0,1])
    print('Sensitivity : ', sensitivity1 )

    specificity1 = cm1[1,1]/(cm1[1,0]+cm1[1,1])
    print('Specificity : ', specificity1)
    acc.append(accuracy1)

# mean_acc = sum(acc)/len(acc)
# mean_sens = sum(sens)/len(sens)
# mean_spec = sum(spec)/len(spec)
print(max(acc))
```
